### Phase 2 – Cleaning


Remove invalid foreign key

```python
sales_valid = sales.join(customers, "customer_id", "inner")
sales_valid.display()
```

```python
cars_clean = cars.filter(col("price") > 0)
cars_clean.display()
```

```python
joined = sales_valid.join(cars_clean, "car_id", "inner")
joined.display()
```

```python
joined = joined.withColumn("revenue" , col("price")*col("quantity"))
joined.display()
```

### Phase 4 – Transformation


1.customer revenue

```python
customer_revenue = joined.groupBy("customer_id", "name").agg(sum("revenue").alias("total_revenue"))

customer_revenue.display()
```

2 Brand-wise sales


```python

brand_sales = joined.groupBy("brand").agg(sum("revenue").alias("Brand_wise_revenue"))
brand_sales.display()
```

3 City-wise revenue


```python

city_sales = joined.groupBy("city").agg(sum("revenue").alias("city_wise_revenue"))
city_sales.display()
```

### Phase 5 – Dealer Analytics
1 Revenue per dealer


```python

dealers = dealers.withColumnRenamed("name", "dealer_name")
dealers = dealers.withColumnRenamed("city", "dealer_city")
```

```python

joined_dealer = joined.join(sales_dealer, "sale_id", "inner").join(dealers, "dealer_id", "inner")
joined_dealer.display()
```

```python
dealer_revenue = joined_dealer.groupBy("dealer_id", "dealer_name").agg(sum("revenue").alias("dealer_wise_revenue"))
dealer_revenue.display()
```

Top dealers by revenue

```python
Top3_dealers = dealer_revenue.orderBy("dealer_wise_revenue", ascending=False).limit(3)
Top3_dealers.display()
```

Dealer-city contribution

```python

dealer_city_contribution = joined_dealer.groupBy("dealer_name","dealer_city").agg(sum("revenue").alias("d_city_revenue"))
dealer_city_contribution.display()
     
```

### SQL Analysis

```python
joined_dealer.createOrReplaceTempView("sales_data")
```

Top 3 customers per city

```python
spark.sql(""" SELECT *
FROM (
    SELECT 
        customer_id,
        city,
        SUM(revenue) AS total_revenue,
        ROW_NUMBER() OVER (
            PARTITION BY city 
            ORDER BY SUM(revenue) DESC
        ) AS rank
    FROM sales_data
    GROUP BY customer_id, city
) t
WHERE rank <= 3 """).show()
```

Monthly trends

```python
spark.sql("""
    SELECT 
        month(sale_date) as month,
        SUM(revenue) AS monthly_revenue
    FROM sales_data
    GROUP BY month
""").show()
     
```

Repeat customers

```python
spark.sql("""
    select customer_id, count(*) as cnt from sales_data group by customer_id having cnt > 1
""").show()
```
