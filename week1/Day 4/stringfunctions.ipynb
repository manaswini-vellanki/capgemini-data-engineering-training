```sql
--Question1

select initcap(emp_name), sum(base_salary+coalesce(bonus,0)) as total_income, round(sum(base_salary+coalesce(bonus,0))) as round_total, year(joining_date)
from employee_payments
group by emp_name, joining_date
```

```sql
select emp_name,
year(curdate())-year(joining_date) as experience,
case
  when year(curdate())-year(joining_date) > 7 then 'Senior'
  when year(curdate())-year(joining_date) between 4 and 7 then 'Mid'
  else 'Junior'
end as emp_level
from employee_payments
```

```sql
--QUESTION 5: Loan EMI Risk Categorization

SELECT 
    loan_id,

    -- Uppercase customer name
    UPPER(customer_name) AS customer_name,

    loan_amount,
    interest_rate,
    loan_start,

    -- 1. Years since loan start
    TIMESTAMPDIFF(YEAR, loan_start, CURRENT_DATE) AS years_since_start,

    -- 2. Monthly interest rate using power logic
    POWER(1 + (interest_rate/100), 1/12) - 1 AS monthly_interest_rate,

    -- 3. EMI Calculation (Rounded)
    ROUND(
        loan_amount * 
        ( (POWER(1 + (interest_rate/100)/12, 12 * 5) * (interest_rate/100)/12) 
        / (POWER(1 + (interest_rate/100)/12, 12 * 5) - 1) )
    , 0) AS emi,

    -- 4. Risk Category
    CASE 
        WHEN interest_rate > 9 THEN 'High Risk'
        WHEN interest_rate BETWEEN 8 AND 9 THEN 'Medium Risk'
        ELSE 'Low Risk'
    END AS risk_category

FROM loan_details;
```
