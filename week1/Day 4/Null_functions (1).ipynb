ISNULL

```sql
--Replace NULL salary with 0

select *,IFNULL(salary,0) as salary from Employees3
```

```sql
--Replace NULL bonus with 1000
select *,IFNULL(bonus,1000) from employees3
```

```sql
--Show order amount, if NULL replace with 500
select *,IFNULL(amount,500) from orders
```

```sql
--Replace NULL stock with 0
select *,IFNULL(stock,0) from products
```

COALESCE

```sql
--Show employee earnings using:
-- salary, if NULL use bonus
select *,coalesce(salary,bonus) from employees3
```

```sql

--Show product price:
 --price → 1000 (default)
 select * ,coalesce(price,1000) from products
```

```sql
--Get customer payment:
 --amount → discount → 0
 select *,coalesce(amount,discount,0) from orders
```

NULLIF

```sql
--Convert salary to NULL if salary = 0
select *,IFNULL(salary,0) as u_s ,NULLIF(u_s,0) from employees3
```

```sql
--Convert discount to NULL if discount = 0
select *,IFNULL(discount,0) as u_d,NULLIF(u_d,0) from orders
```

```sql
SELECT *,amount / NULLIF(discount, 0)
FROM Orders;
```

```sql
--Replace coupon_code with NULL if it is 'DISC10'
select * , NULLIF(coupon_code,'DISC10')from orders
```

```sql
--Calculate total earnings:
   --salary + bonus (handle NULL properly)
   SELECT name,
       COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM employees3;
```

```sql
--Replace empty discount with NULL using NULLIF:
select  *,NULLIF(discount, 0) from orders
```

```sql
--Show final payable amount:
 -- amount - discount (handle NULL)
select *,coalesce(amount,0)-coalesce(discount,0) from orders
```
