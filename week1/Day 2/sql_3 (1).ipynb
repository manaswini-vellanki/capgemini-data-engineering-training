###Basic GROUP BY Queries

```sql
SELECT department,sum(salary) as total_salary
FROM employee department 
GROUP BY department;
```

```sql
select department,count(*) as total_count
from employee
group by department
```

```sql
select department,avg(salary) as avg_salary
from employee
group by department
```

```sql
select department,max(salary)
from employee
group by department
```

```sql
select department,min(salary)
from employee
group by department
```

###GROUP BY with Conditions

```sql
select department,sum(salary)
from employee
group by department
having sum(salary)>100000
```

```sql
select department,count(*)
from employee
group by department
having count(*)>2
```

```sql
select department,avg(Salary)
from employee
where joining_date > '2021-01-01'
group by department
```

```sql
select department,max(Salary) as max_salary
from employee
group by department
having max_salary >75000
```

```sql
select department,sum(salary) as total_salary
from employee
group by department
having (total_salary) <150000
```

###GROUP BY with HAVING

```sql
select department,count(*)
from employee
group by department
having count(*)>1
```

```sql
select department,sum(salary) as total_salary
from employee
group by department
having sum(salary)>125000
```

```sql
select department,count(*) as total_employees
from employee
group by department
having (total_employees)>2
```

```sql
select department,avg(salary) as avg_salary
from employee
group by department
having avg(salary)>60000
```

```sql
select department,sum(salary) as total_salary
from employee
group by department
having sum(salary)  between 100000 and 200000
```

```sql
select e.emp_name,sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id=s.emp_id
group by e.emp_name
```

```sql
select e.emp_name,count(*) as total_sales
from employee e
left join sales s
on e.emp_id=s.emp_id
group by e.emp_name
```

```sql
select s.emp_id,s.product,count(s.amount)
from sales s
join employee e
on s.emp_id=e.emp_id
group by s.product,s.emp_id
```

```sql
select s.emp_id,s.product,avg(s.amount)
from sales s
join employee e
on s.emp_id=e.emp_id
group by product,s.emp_id
```

```sql
select e.emp_name,count(s.emp_id) as count_sales
from sales s
join employee e
on s.emp_id=e.emp_id
group by e.emp_name
having count_sales>2
```

```sql
select e.emp_id,sum(e.salary),sum(s.amount)
from sales s
join employee e
on e.emp_id=s.emp_id
group by e.emp_id
```

```sql
select emp_id,count(distinct product)
from sales
group by emp_id
```

```sql
select emp_id,max(amount)
from sales
group by emp_id
```

```sql
select product,sum(amount) as total
from sales
group by product
having total>50000
```

```sql
select e.department,max(s.amount) as highest_amount
from employee e
join sales s
on e.emp_id=s.emp_id
group by e.department
```

```sql
select e.department,sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id=s.emp_id
group by e.department 
order by total_sales desc
limit 1
```

```sql
select e.emp_name,sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id=s.emp_id
group by e.emp_name
order by total_sales desc
limit 3
```

```sql
select count(e.emp_id),year(joining_date) as year_joining,avg(e.salary) as average_salary
from employee e
group by year_joining
```

```sql
select e.department ,sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id=s.emp_id
group by e.department
```
