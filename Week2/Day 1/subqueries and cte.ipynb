```sql
CREATE  OR REPLACE TABLE Employees (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);

INSERT INTO Employees VALUES
(1, 'Alice', 101, 60000, NULL),
(2, 'Bob', 102, 45000, 1),
(3, 'Charlie', 101, 70000, 1),
(4, 'David', 103, 40000, 2),
(5, 'Eve', 102, 50000, 2);
```

```sql
CREATE  OR REPLACE TABLE Departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);

INSERT INTO Departments VALUES
(101, 'HR'),
(102, 'IT'),
(103, 'Finance');
```

```sql
CREATE OR REPLACE TABLE Orders (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);

INSERT INTO Orders VALUES
(1, 1, 5000, '2024-01-10'),
(2, 2, 3000, '2024-02-15'),
(3, 3, 7000, '2024-03-12'),
(4, 1, 2000, '2024-04-01'),
(5, 4, 1000, '2024-05-05');
```

```sql
SELECT *
FROM Employees
WHERE salary > (
    SELECT MAX(salary)
    FROM Employees
    WHERE department_id = 102
);
```

```sql
with high_salary_emp as (select * 
from employees
where salary>50000)
select * from high_salary_emp
```

```sql
with dept_avg as(
  select department_id,avg(Salary) as avg_salary
  from employees
  group by department_id 
)
select * from dept_avg
```

```sql
with emp_names as(
  select e.emp_id,d.dept_name
  from employees e
  join departments d on e.department_id=d.dept_id
)
select * from emp_names
```

```sql
with total_amount as(
  select emp_id,sum(order_amount) as total
  from orders
  group by emp_id
  
)
select * from total_amount
```

```sql
WITH dept_avg AS (
    SELECT department_id, AVG(salary) AS avg_salary
    FROM Employees
    GROUP BY department_id
)
SELECT e.*
FROM Employees e
JOIN dept_avg d
ON e.department_id = d.department_id
WHERE e.salary > d.avg_salary;
```
